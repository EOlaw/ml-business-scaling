```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random
```

```python
# Set random seed for reproducibility
np.random.seed(42)
```

```python
# --- USERS DATASET ---
num_users = 1000
user_ids = range(1001, 1001 + num_users)
ages = np.random.randint(18, 60, num_users)
genders = np.random.choice(["M", "F"], num_users)
regions = np.random.choice(["North", "South", "East", "West", "Midwest"], num_users)
signup_dates = [datetime(2024, 1, 1) + timedelta(days=np.random.randint(0, 180)) for _ in range(num_users)]
referral_sources = np.random.choice(["Google Ads", "Referral", "Instagram", "LinkedIn", "Facebook"], num_users)

users = pd.DataFrame({
    "user_id": user_ids,
    "age": ages,
    "gender": genders,
    "region": regions,
    "signup_date": signup_dates,
    "referral_source": referral_sources
})
print(users)
```

```python
# --- CONSULTATIONS DATASET ---
num_consultations = 2000
consultation_ids = range(5001, 5001 + num_consultations)
user_id_choices = np.random.choice(user_ids, num_consultations)
consultant_ids = np.random.randint(200, 250, num_consultations)
dates = [datetime(2024, 3, 1) + timedelta(days=np.random.randint(0, 180)) for _ in range(num_consultations)]
durations = np.random.choice([30, 45, 60, 90], num_consultations, p=[0.3, 0.3, 0.3, 0.1])
prices = np.random.choice([40, 60, 80, 120], num_consultations, p=[0.25, 0.35, 0.3, 0.1])
completed = np.random.choice([True, False], num_consultations, p=[0.75, 0.25])

consultations = pd.DataFrame({
    "consultation_id": consultation_ids,
    "user_id": user_id_choices,
    "consultant_id": consultant_ids,
    "date": dates,
    "duration": durations,
    "price": prices,
    "completed": completed
})
print(consultations)
```

```python
# --- MARKETING DATASET ---
channels = ["Google Ads", "Instagram", "LinkedIn", "Facebook", "Referral"]
num_campaigns = 50
marketing = pd.DataFrame({
    "campaign_id": range(1, num_campaigns + 1),
    "channel": np.random.choice(channels, num_campaigns),
    "spend": np.random.randint(500, 5000, num_campaigns),
    "clicks": np.random.randint(100, 2000, num_campaigns),
    "conversions": np.random.randint(10, 400, num_campaigns),
    "date": [datetime(2024, 4, 1) + timedelta(days=i * 7) for i in range(num_campaigns)]
})
print(marketing)
```

```python
# --- SAVE FILES ---
users.to_csv("users.csv", index=False)
consultations.to_csv("consultations.csv", index=False)
marketing.to_csv("marketing.csv", index=False)

"users.csv", "consultations.csv", "marketing.csv"
```
